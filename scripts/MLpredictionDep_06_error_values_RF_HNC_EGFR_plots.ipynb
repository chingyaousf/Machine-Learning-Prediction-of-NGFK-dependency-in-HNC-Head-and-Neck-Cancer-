```python
# RF_HNC_EGFR_error_values.csv

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file for EGFR
df_egfr = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HNC/RF_HNC_EGFR_error_values.csv')

# Extract the required columns for EGFR
k_means_cluster = df_egfr['K-Means cluster']
mse = df_egfr['MSE']
mae = df_egfr['MAE']
r2 = df_egfr['R2']

# Bar plot
plt.figure(figsize=(10, 6))
plt.bar(range(len(k_means_cluster)), mse, width=0.2, align='center', label='MSE')
plt.bar([i + 0.2 for i in range(len(k_means_cluster))], mae, width=0.2, align='center', label='MAE')
plt.bar([i + 0.4 for i in range(len(k_means_cluster))], r2, width=0.2, align='center', label='R2')
plt.xlabel('K-Means Cluster')
plt.ylabel('Metric Value')
plt.title('EGFR Gene: Metrics by K-Means Cluster')
plt.legend()
plt.tight_layout()

# Set the k-means cluster labels as tick labels
tick_labels = [f'Cluster {i}' for i in k_means_cluster]
plt.xticks(range(len(k_means_cluster)), tick_labels)
plt.gca().tick_params(axis='x', pad=0)  # Adjust the position of x-axis labels

plt.show()
```
